```python
#DCPP Assignment- Group No: 4, Section A
#Exploratory Data Analysis of BSE listed Companies 
#importing all related packages
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import datetime
Companydata= pd.read_csv("C:/Users/mahes/Desktop/DCPP/BSE Trading View.csv")
```

```python
#checking whether data is loaded properly or not
Companydata.head(10)
```

```python
#Understanding aspects of BSE listed Companies data 
Companydata.describe()
```

```python
import matplotlib.pyplot as plt
```

```python
#Using Pivot Table for plotting stock price basis sector 

pivot=Companydata.pivot_table(index=['Sector'],values=['Stock Price'], aggfunc={'sum','median', 'mean'})
print(pivot)
```

```python
#Using Pivot Table for plotting market capitalization basis sector 

pivot=Companydata.pivot_table(index=['Sector'],values=['Market Capitalization'], aggfunc={'sum','median', 'mean'})
print(pivot)
```

```python
#Using Pivot Table for plotting volumes traded basis sector 

pivot=Companydata.pivot_table(index=['Sector'],values=['Volume'], aggfunc={'sum','median', 'mean'})
print(pivot)
```

```python
# Plotting of sector wise market capitalisation 
Companydata1 = Companydata.groupby('Sector').sum()
plt.figure(figsize=(50, 20))
plt.title('Sector wise market capitalisation')
plt.xlabel('Sector')
plt.ylabel('Market Capitalization')

plt.bar(Companydata1.index, Companydata1['Market Capitalization'])
plt.show()
```

```python
# Plotting of sector wise Volumes  

Companydata1 = Companydata.groupby('Sector').sum()
plt.figure(figsize=(50, 20))
plt.title('Sector wise Volumes    ')
plt.xlabel('Sector')
plt.ylabel('Volume')

plt.bar(Companydata1.index, Companydata1['Volume'])
plt.show()
```
